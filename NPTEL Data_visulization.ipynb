```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('iris.csv',index_col=0)
```

```python
plt.scatter(data['Sepal.Length'],data['Petal.Length'],c='red')
plt.title('scatter plot')
plt.xlabel('Sepal_Length')
plt.ylabel('Petal_Length')
plt.show() # To avoid this command, you can add %matplotlib inline in import library block
```

```python
plt.hist(data['Sepal.Length'],color='green',edgecolor='white',bins=5)
plt.title('Histogram')
plt.xlabel('Sepal.Length')
plt.ylabel('Fequecy)
plt.show()
```

```python
counts=[979,120,12]
fueltype=['petrol','disel','cng']
index=np.arange(len(fueltype))
```

```python
plt.bar(index,counts,color=['red','blue','cyan',])
plt.title('Bar plot of fuel types')
plt.xlabel('Fuel types')
plt.ylabel('Frequency')
plt.xticks(index,fueltype,rotation=90)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('hart.csv')
```

```python
sns.set(style="darkgrid")
```

```python
sns.regplot(x=data['chol'],y=data['thalach'],marker="*",fit_reg=False)
```

```python
sns.lmplot(x='chol',y='thalach',data=data,fit_reg=False,hue='age',palette="Set1",legend=True)
```

```python
sns.distplot(data['chol'])
```

```python
sns.distplot(data['chol'],kde=False,bins=5)
```

```python
sns.countplot("sex",data=data)
```

```python
sns.countplot("sex",data=data,hue="num")
```

```python
pd.crosstab(index=data['num'],columns=['sex'])
```

```python
sns.boxplot(data['chol'])
```

```python
sns.boxplot(x=data['chol'],y=data['sex'])
```

```python
sns.boxplot(data['chol'],data['sex'],hue="num",data=data)
```

```python
sns.pairplot(data,kind="scatter",hue="num")
plt.show()
```
